In this task you will use [Air Quality Index (AQI)](https://www.airnow.gov/aqi/aqi-basics/) data from the [US Environmental Protection Agency](https://www.epa.gov) to visualize the impact on the AQI of the 2017 [Thomas Fire](https://en.wikipedia.org/wiki/Thomas_Fire) in Santa Barbara County.

```python
# Import the pandas library as pd
import pandas as pd
# Import the matplotlib library as plt
import matplotlib.pyplot as plt


# Read in data
aqi_17 = pd.read_csv('https://aqs.epa.gov/aqsweb/airdata/daily_aqi_by_county_2017.zip', compression='zip')
aqi_18 = pd.read_csv('https://aqs.epa.gov/aqsweb/airdata/daily_aqi_by_county_2018.zip', compression='zip')

# Concatenate the aqi_17 and aqi_18 DataFrames into one
aqi = pd.concat([aqi_17, aqi_18])

# Simplify column names
aqi.columns = (aqi.columns
                  .str.lower()
                  .str.replace(' ','_'))

# Select the rows where the country_name is Santa Barbara and assign them too aqi_sb
aqi_sb = aqi[aqi['county_name'] == "Santa Barbara"]

# Remove the 'state_name', 'county_name', 'state_code', and 'county_code' columns from the aqi_sb DataFrame
aqi_sb = aqi_sb.drop(columns = ['state_name', 'county_name', 'state_code', 'county_code'])

# Change the date column data type from an object to datetime
aqi_sb.date = pd.to_datetime(aqi_sb["date"])

# Set the date column as the index
aqi_sb = aqi_sb.set_index('date')

# Calculate AQI rolling average over 5 days
rolling_average = aqi_sb['aqi'].rolling('5D').mean()

# Add the 5-day rolling average from `rolling_average` to a new column named `five_day_average` to the `aqi_sb` dataframe.
aqi_sb['five_day_average'] = rolling_average

# Plot the both the daily AQI and the 5-day average of AQI from the aqi_sb DataFrame
aqi_sb.plot(kind="line",
            y = ["aqi", 'five_day_average'],
            title = "Air Quality index in Santa Barbara Country ()",
            xlabel = "Date",
            ylabel = "Air Quality Index",
            color = ['hotpink', 'blue'])
plt.suptitle("Thomas Fire (12/17) ->", fontsize=10, y=0.80, x=.35)  # Edit font size and position of subtitle
plt.show()
```

- `rolling()` is a lazy method, so we need to specify what we want to calculate over each window before it does something. 
- in this example we use the aggregator function `mean()` to calculate the average over each window
- the parameter '5D' indicates we want the window for our rolling average to be 5 days.

Store your answer in the `rolling_average` variable. You should have two columns in your series, `date` and the averages.
